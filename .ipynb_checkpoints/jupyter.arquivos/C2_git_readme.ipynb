## Algoritmo para Verificar se uma S√©rie √© Geom√©trica

# üìò Teste do Termo n-√©simo

O **Teste do Termo n-√©simo** √© uma ferramenta fundamental no estudo de **s√©ries infinitas**.

---

## üîπ A Regra Fundamental

Para que a s√©rie

$\sum_{n=1}^{\infty} a_n$

$\lim_{n \to \infty} a_n = 0$

---

## üí° Intui√ß√£o

Imagine que voc√™ est√° tentando encher um copo (o limite $S$) com infinitas gotas (os termos $a_n$):

- **Se as gotas nunca diminuem de tamanho:**  
  Se $a_n$ n√£o tende a zero (por exemplo, $a_n \to 0.5$), voc√™ estar√° adicionando $0.5 + 0.5 + 0.5 + \dots$.  
  O copo nunca para de encher ‚Üí **a s√©rie diverge**.

- **Se as gotas ficam cada vez menores:**  
  A √∫nica chance de converg√™ncia √© se $a_n \to 0$.  
  Nesse caso, eventualmente, voc√™ estar√° adicionando "quase nada" ao copo.

---

## üìê Demonstra√ß√£o Formal (Explicada)

O argumento √© feito usando **somas parciais**.

1. **Defini√ß√µes**
   - $S$: o limite da s√©rie (soma total).  
   - $s_n$: a soma parcial at√© o termo $n$.  
     $s_n = a_1 + a_2 + \dots + a_n$  
   - $s_{n-1}$: a soma parcial anterior.  

2. **Rela√ß√£o chave**
   $$a_n = s_n - s_{n-1}$$

3. **C√°lculo do limite**
   Se a s√©rie converge para $S$, ent√£o:
   $$\lim_{n \to \infty} s_n = S \quad \text{e} \quad \lim_{n \to \infty} s_{n-1} = S$$

   $$\lim_{n \to \infty} a_n = \lim_{n \to \infty}(s_n - s_{n-1}) = S - S = 0$$

‚úÖ Conclus√£o: **se a s√©rie converge, ent√£o obrigatoriamente $a_n \to 0$.**

---

## üìä Exemplo Pr√°tico (do Livro)

Considere a s√©rie:

$\sum_{n=1}^{\infty} \frac{n}{n+1} = \frac{1}{2} + \frac{2}{3} + \frac{3}{4} + \dots$

- Termo geral:
  $$a_n = \frac{n}{n+1}$$

- Limite do termo:
  $$\lim_{n \to \infty} \frac{n}{n+1} = \lim_{n \to \infty} \left( 1 - \frac{1}{n+1} \right) = 1$$

- Como o limite $\neq 0$, o teste garante que:
  $$\text{A s√©rie DIVERGE}$$

üîé Intui√ß√£o: no fim das contas, voc√™ est√° somando muitos n√∫meros que s√£o, essencialmente, iguais a 1.  
E $1 + 1 + 1 + \dots$ infinito ‚Üí **diverge**.

---

## ‚ö†Ô∏è Observa√ß√£o Importante

O Teste do Termo n-√©simo √© √∫til principalmente para **provar diverg√™ncia**:

- Se $\lim_{n \to \infty} a_n \neq 0$ ‚Üí a s√©rie **diverge**. ‚úÖ  
- Se $\lim_{n \to \infty} a_n = 0$ ‚Üí o teste √© **inconclusivo**. ‚ùå  

Exemplo cl√°ssico:  
$$\sum_{n=1}^{\infty} \frac{1}{n}$$
Aqui $a_n = \frac{1}{n} \to 0$, mas a s√©rie **diverge** (S√©rie Harm√¥nica).  

Portanto, o teste do termo n-√©simo **n√£o prova converg√™ncia**, apenas diverg√™ncia.

## Fator Integrante ‚Äî Explica√ß√£o Completa

O **fator integrante** √© uma t√©cnica cl√°ssica para resolver **equa√ß√µes diferenciais lineares de 1¬™ ordem** do tipo:

$$
y' + P(x)y = Q(x)
$$

onde $P(x)$ e $Q(x)$ s√£o fun√ß√µes conhecidas de $x$.

---

### 1. Forma padr√£o

Antes de aplicar o m√©todo, √© fundamental escrever a EDO nesta forma:

$$
y' + P(x)y = Q(x)
$$

**Exemplo:**

$$
y' + 2y = e^{-x}
$$

Aqui:
- $P(x) = 2$
- $Q(x) = e^{-x}$

---

### 2. Ideia do fator integrante

Queremos multiplicar toda a equa√ß√£o por uma fun√ß√£o $\mu(x)$ (chamada **fator integrante**) de modo que o lado esquerdo vire uma **derivada de um produto**:

$$
mu(x)y' + \mu(x)P(x)y = \frac{d}{dx}[\mu(x)y]
$$

---

### 3. Como encontrar $\mu(x)$

Pela **regra do produto**, sabemos que:

$$
frac{d}{dx}[\mu(x)y] = \mu'(x)y + \mu(x)y'
$$

Comparando com o lado esquerdo da EDO multiplicada ($\mu(x)y' + \mu(x)P(x)y$), percebemos que os termos ser√£o id√™nticos **se**:

$$
mu'(x) = \mu(x)P(x)
$$

---

### 4. Resolvendo para $\mu(x)$

Essa √© uma EDO separ√°vel:

$$
\frac{\mu'(x)}{\mu(x)} = P(x)
$$

Separando e integrando ambos os lados:

$$
\int \frac{\mu'(x)}{\mu(x)}dx = \int P(x)\,dx
$$

Sabemos pela **regra da derivada do logaritmo natural** que:

$$
\frac{d}{dx}[\ln(\mu)] = \frac{\mu'(x)}{\mu(x)}
$$

Logo, ao integrar, obtemos:

$$
ln|\mu| = \int P(x)\,dx + C
$$

Exponenciando ambos os lados:

$$
mu(x) = e^{\int P(x)\,dx + C} = e^{C} \cdot e^{\int P(x)\,dx}
$$

Como $e^C$ √© uma constante multiplicativa, podemos ignor√°-la (ou tom√°-la igual a 1):

$$
\boxed{\mu(x) = e^{\int P(x)\,dx}}
$$

---

### 5. Aplicando o fator integrante

Multiplicamos toda a EDO original por $\mu(x)$:

$$
\mu(x)y' + \mu(x)P(x)y = \mu(x)Q(x)
$$

O lado esquerdo agora √© a **derivada de um produto**:

$$
\frac{d}{dx}[\mu(x)y] = \mu(x)Q(x)
$$

Integrando ambos os lados:

$$
\mu(x)y = \int \mu(x)Q(x)\,dx + C
$$

E finalmente isolamos $y(x)$:

$$
\boxed{y(x) = \frac{1}{\mu(x)}\left(\int \mu(x)Q(x)\,dx + C\right)}
$$

---

### 6. Exemplo completo

Dada a EDO:

$$
y' + 2y = e^{-x}
$$

Temos $P(x) = 2$ e $Q(x) = e^{-x}$.

1. **Calcula o fator integrante:**

$$
\mu(x) = e^{\int 2dx} = e^{2x}
$$

2. **Multiplica a EDO por $\mu(x)$:**

$$
e^{2x}y' + 2e^{2x}y = e^{x}
$$

3. **Reconhece a derivada de um produto:**

$$
frac{d}{dx}(e^{2x}y) = e^{x}
$$

4. **Integra ambos os lados:**

$$
e^{2x}y = \int e^{x}\,dx = e^{x} + C
$$

5. **Isola $y$:**

$$
y = e^{-x} + Ce^{-2x}
$$

---

### 7. Regra geral da derivada do logaritmo natural

Se $f(x)$ √© uma fun√ß√£o deriv√°vel e **positiva**, ent√£o:

$$
\boxed{\frac{d}{dx}[\ln(f(x))] = \frac{f'(x)}{f(x)}}

> Essa regra vale para **qualquer fun√ß√£o diferenci√°vel positiva** $f(x)$.

Vers√£o mais geral (v√°lida mesmo se $f(x) < 0$):

$$

### Intui√ß√£o

- $\ln(x)$ tem derivada $1/x$.
- $\ln(f(x))$ √© a composi√ß√£o de $\ln$ com $f(x)$, ent√£o usamos a **regra da cadeia**:

\frac{d}{dx}[\ln(f(x))] = \frac{1}{f(x)} \cdot f'(x)

---

#### Exemplos

$$

$$

3. $f(x) = e^{3x} + 5$

# Prova do Teste da Raz√£o

## üìã Enunciado do Teorema

**Teste da Raz√£o:** Seja $\sum a_n$ uma s√©rie com termos positivos. Se

$$\lim_{n \to \infty} \frac{a_{n+1}}{a_n} = \rho$$

ent√£o:
- Se $\rho < 1$, a s√©rie **converge**
- Se $\rho > 1$, a s√©rie **diverge**
- Se $\rho = 1$, o teste √© **inconclusivo**

---

## ‚úÖ Demonstra√ß√£o do Caso: $\rho < 1$

### Passo 1: Escolha de $r$ e $\varepsilon$

Como $\rho < 1$, podemos escolher um n√∫mero $r$ tal que:

$$\rho < r < 1$$

Definimos ent√£o $\varepsilon = r - \rho > 0$, de modo que:

$$\rho + \varepsilon = r < 1$$

> **Objetivo:** Usar $r$ como raz√£o de uma s√©rie geom√©trica convergente que "domina" nossa s√©rie.

---

### Passo 2: Aplica√ß√£o da Defini√ß√£o de Limite

Como $\displaystyle \lim_{n \to \infty} \frac{a_{n+1}}{a_n} = \rho$, pela defini√ß√£o de limite:

Para o $\varepsilon > 0$ escolhido, existe $N \in \mathbb{N}$ tal que para todo $n \geq N$:

$$\frac{a_{n+1}}{a_n} < \rho + \varepsilon = r$$

> **Interpreta√ß√£o:** A partir de um certo √≠ndice $N$, a raz√£o entre termos consecutivos fica consistentemente menor que $r$.

---

### Passo 3: Constru√ß√£o das Desigualdades em Cadeia

Aplicando repetidamente a desigualdade $\frac{a_{n+1}}{a_n} < r$ para $n \geq N$:

$$\begin{align}
a_{N+1} &< r \cdot a_N \\
a_{N+2} &< r \cdot a_{N+1} < r^2 \cdot a_N \\
a_{N+3} &< r \cdot a_{N+2} < r^3 \cdot a_N \\
&\vdots \\
a_{N+m} &< r \cdot a_{N+m-1} < r^m \cdot a_N
\end{align}$$

> **Visualiza√ß√£o:** Cada termo ap√≥s $a_N$ √© limitado superiormente por m√∫ltiplos de pot√™ncias de $r$.

---

### Passo 4: Constru√ß√£o da S√©rie Auxiliar

Definimos uma s√©rie auxiliar $\sum c_n$ onde:

$$c_n = \begin{cases}
a_n & \text{se } n = 1, 2, \ldots, N \\
r^{n-N} \cdot a_N & \text{se } n > N
\end{cases}$$

Ent√£o:

$$\begin{align}
\sum_{n=1}^{\infty} c_n &= a_1 + a_2 + \cdots + a_{N-1} + a_N + ra_N + r^2a_N + r^3a_N + \cdots \\
&= \underbrace{a_1 + a_2 + \cdots + a_{N-1}}_{\text{soma finita}} + a_N \underbrace{(1 + r + r^2 + r^3 + \cdots)}_{\text{s√©rie geom√©trica}}
\end{align}$$

---

### Passo 5: Converg√™ncia da S√©rie Geom√©trica

A s√©rie geom√©trica $1 + r + r^2 + r^3 + \cdots$ converge porque $|r| < 1$, e sua soma √©:

$$\sum_{k=0}^{\infty} r^k = \frac{1}{1-r}$$

Portanto:

$$\sum_{n=1}^{\infty} c_n = (a_1 + a_2 + \cdots + a_{N-1}) + a_N \cdot \frac{1}{1-r} < \infty$$

> **Conclus√£o Parcial:** A s√©rie $\sum c_n$ **converge**.

---

### Passo 6: Teste da Compara√ß√£o

Por constru√ß√£o, temos que:

$$a_n \leq c_n \quad \text{para todo } n \geq 1$$

Como:
1. Todos os termos s√£o positivos
2. $\sum c_n$ converge
3. $a_n \leq c_n$ para todo $n$

Pelo **Teste da Compara√ß√£o**, conclu√≠mos que:

$$\boxed{\sum_{n=1}^{\infty} a_n \text{ converge}}$$

---

## üéØ Intui√ß√£o Geom√©trica

A prova essencialmente mostra que:

- Ap√≥s um certo ponto ($n \geq N$), nossa s√©rie se comporta **melhor** que uma s√©rie geom√©trica com raz√£o $r < 1$
- Como a s√©rie geom√©trica converge, nossa s√©rie tamb√©m deve convergir
- A escolha de $r$ entre $\rho$ e $1$ garante que temos uma "margem de seguran√ßa" para a converg√™ncia

---

## üìä Exemplo Visual

Considere $\rho = 0.5$ e escolhemos $r = 0.7$:

| Posi√ß√£o | S√©rie Original | | S√©rie Limitante (geom√©trica) |
|:-------:|:--------------:|:---:|:---------------------------:|

**Conclus√£o:** Como $r = 0.7 < 1$, a s√©rie geom√©trica $\sum (0.7)^m a_N$ converge, for√ßando a converg√™ncia da s√©rie original $\sum a_n$! ‚úì
